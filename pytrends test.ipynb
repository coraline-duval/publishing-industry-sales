```python
import pytrends
from pytrends.request import TrendReq
import pandas as pd
import json
```

```python
f = open('titles_book_id',)
titles = json.load(f)
```

```python
pytrend = TrendReq(hl='en-US', tz=360)
res=pd.DataFrame()

startpoint=35460

for i in range(len(titles[startpoint:])//5 +1): 
    temp=pd.DataFrame() #initialising so it doesn't fill in result from before
    kw_list=[list(i.keys())[0] for i in titles[startpoint:][i*5:i*5 + 5]]
    pytrend.build_payload(kw_list, cat=22, timeframe='2010-01-01 2020-12-01') #books and litterature is cat 22, selecting timeframe after 2010 as only books published after 2010
    try:
        temp=pytrend.interest_over_time()[kw_list].transpose()
        temp['book_id']=[list(i.values())[0] for i in titles[startpoint:][i*5:i*5 + 5]]
        res=pd.concat([res,temp])
    except : 
        print(titles[startpoint:][i*5:i*5 + 5])
    
```
